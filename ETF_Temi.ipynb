```python
import pandas as pd
```

```python
Oscar_file = "Oscars.csv"
oscars_df = pd.read_csv(Oscar_file, encoding='iso-8859-1')
oscars_df.head()
```

```python
# Create a filtered dataframe from specific columns
oscar_cols= ["movie","award", "birthplace","date_of_birth"]
oscar_transformed= oscars_df[oscar_cols].copy()

# Rename the column headers
oscar_transformed = oscar_transformed.rename(columns={"award": "Award",
                                                      "movie":"Movies",
                                                      "birthplace": "Film Birth Place"
                                                     })

# Clean the data by dropping duplicates and setting the index
oscar_transformed.drop_duplicates("Movies", inplace=True)
oscar_transformed.set_index("Movies", inplace=True)

oscar_transformed.head(40)
```
