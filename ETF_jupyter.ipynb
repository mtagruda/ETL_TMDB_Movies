```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
movie_file = "tmdb_5000_movies.csv"
movies_df = pd.read_csv(movie_file)
movies_df.head()

```

```python
Oscar_file = "Oscars.csv"
oscars_df = pd.read_csv(Oscar_file, encoding='iso-8859-1')
oscars_df.head()
```

```python
# Create a filtered dataframe from specific columns
movie_cols= ["homepage","release_date", "title", "popularity"]
movie_transformed= movies_df [movie_cols].copy()
# Rename the column headers
movie_transformed = movie_transformed.rename(columns={"Homepage": "Homepage",
                                                     "release_date":"Release Date",
                                                     "title": "title"
                                                    })
# Clean the data by dropping duplicates and setting the index
movie_transformed.drop_duplicates("title", inplace=True)
movie_transformed.set_index("title", inplace=True)
movie_transformed.head(30)
```

```python
# Create a filtered dataframe from specific columns
movies_cols = ["License Serial Number", "Premises Name", "County ID Code"]
movies_transformed= movies_df[movies_cols].copy()

# Rename the column headers
premise_transformed = premise_transformed.rename(columns={"License Serial Number": "id",
                                                          "Premises Name": "premise_name",
                                                          "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
premise_transformed.drop_duplicates("id", inplace=True)
premise_transformed.set_index("id", inplace=True)

premise_transformed.head()
```

```python
browser = Browser('chrome', headless = False)
```

```python
# URL to be scrapped
browser.visit('https://www.rottentomatoes.com/browse/dvd-streaming-all')
```

```python
for i in range(1,100):
    try:    
        browser.find_by_text('Show More').first.click() 
    
    except: 
        print("Scraping Complete")
```

```python
soup = BeautifulSoup(browser.html, 'html.parser')
parent = soup.find_all(class_='mb-movie')
```

```python
movie_titles = [] 
ratings = []


for movie in parent:
    title = movie.find(class_='movieTitle').text
    rating = movie.find(class_='tMeterScore').text
    movie_titles.append(title)
    ratings.append(rating)
```
