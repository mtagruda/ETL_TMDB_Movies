```python
import pandas as pd
from sqlalchemy import create_engine
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
movie_file = "tmdb_5000_movies.csv"
movies_df = pd.read_csv(movie_file)
movies_df.head()
```

```python
# Create a filtered dataframe from specific columns
movie_cols= ["homepage","release_date", "title", "popularity"]
movie_transformed= movies_df [movie_cols].copy()
# Rename the column headers
movie_transformed = movie_transformed.rename(columns={"Homepage": "Homepage",
                                                     "release_date":"Release Date",
                                                     "title": "Movies", "popularity": "Popularity"
                                                    })
# Clean the data by dropping duplicates and setting the index
movie_transformed.drop_duplicates("Movies", inplace=True)
movie_transformed.set_index("Movies", inplace=True)
movie_transformed.head(30)
```

```python
Oscar_file = "Oscars.csv"
oscars_df = pd.read_csv(Oscar_file, encoding='iso-8859-1')
oscars_df.head()
```

```python
# Create a filtered dataframe from specific columns
oscar_cols= ["movie","award", "birthplace","date_of_birth"]
oscar_transformed= oscars_df[oscar_cols].copy()

# Rename the column headers
oscar_transformed = oscar_transformed.rename(columns={"award": "Award",
                                                      "movie":"Movies",
                                                      "birthplace": "Film Birth Place"
                                                     })

# Clean the data by dropping duplicates and setting the index
oscar_transformed.drop_duplicates("Movies", inplace=True)
oscar_transformed.set_index("Movies", inplace=True)

oscar_transformed.head(40)
```

```python
browser = Browser('chrome', headless = False)
```

```python
# URL to be scrapped
browser.visit('https://www.rottentomatoes.com/browse/dvd-streaming-all')
```

```python
for i in range(1,100):
    try:    
        browser.find_by_text('Show More').first.click() 
    
    except: 
        print("Scraping Complete")
```

```python
soup = BeautifulSoup(browser.html, 'html.parser')
rotten = soup.find_all(class_='mb-movie')
```

```python
movie_titles = [] 
ratings = []


for movie in rotten:
    title = movie.find(class_='movieTitle').text
    rating = movie.find(class_='tMeterScore').text
    movie_titles.append(title)
    ratings.append(rating)
```

```python
len(movie_titles)
```

```python
rotten_df = pd.DataFrame(list(zip(movie_titles, ratings)), columns = ['Title', 'Rating'])
rotten_df
```

```python
connection_string = "zach:zach@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
rotten_df.to_sql(name='rotten', con=engine, if_exists='append', index=True)
```

```python
movie_transformed.to_sql(name='movie', con=engine, if_exists='append', index=True)
```

```python
oscar_transformed.to_sql(name='oscar', con=engine, if_exists='append', index=True)
```
